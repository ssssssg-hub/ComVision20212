```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('dsu4.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[235,457],[238,269],[569,231],[564,493]])
pts2 = np.float32([[235,457],[235,270],[570,270],[570,457]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(0,0))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()img = cv2.imread('book.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[350,370],[800,370],[40,880],[760,1000]])
pts2 = np.float32([[0,0],[800,0],[0,900],[800,900]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(958,1276))
plt.rcParams["figure.figsize"] = (20,10)
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
